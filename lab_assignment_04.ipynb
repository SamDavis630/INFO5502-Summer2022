Please conduct the following analysis:

(2)  Visualize the count for each value in the 'Study Hours' column. Note: Study Hours values comes from this >= 6 hours Good >= 4 hours Average < 2 hours Poor.

(3) Using factorplot to show the relationship between numerical and categorical value. Please use figure to show relation between [Study Hours] and [End Semester Percentage]. Percentage is calculated based on the following:

if percentage >=80 then Best
If percentage >= 60 but less than 80 then Very Good
If percentage >= 45 but less than 60 then Good
If Percentage >= 30 but less than 45 then Pass
If Percentage < 30 then Fail

(4) Get the correlation from the data, then plug it to heatmap function and show it as following:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

import io
df1 = pd.read_csv(io.BytesIO(uploaded["lab-assignment-04-student-performance.csv"]))

#Create a dataframe
df2 = df1.copy() 
df3 = df1.copy()

print(df2.corr)


##1 Description of all the columns: for each colmun, count number of rows, number of unqiue element, the frequency of each unqiue element.

#created a loop to describe each variable
for column in df2.columns:
    print("---------------")
    print("Column:",column)
    print("number of rows:",len(df2[column]))
    print("number of unique elements:",len(df2[column].unique()))
    print(f"frequency of each unique element:\n{df2[column].value_counts().to_string()}")
  

##2 Visualize the count of each value in the Study Hours column.
print("Study Hours Rating")
df2['sh'].value_counts().plot(kind = 'bar')
plt.ylabel("Total")
plt.xlabel("Rating")
plt.xticks(ticks = [0,1,2], labels = ["Average", "Good", "Poor"], rotation = 45)
plt.show()

##3 Using factorplot to show the relationship between numerical and categorical value. 
print("Factor Plot") 

#defined and changed sh to a numeric variable
dict_1 = {'Best':'>= 80', 'Vg':'60 <= P <80', 'Good':'45 <= P < 60','Pass':'30 <= P < 45','Fail':'< 30'}
dict_2 = {'Good':3, "Average":2, "Poor":1}

df2['n.esp'] = df2['esp'].map(dict_1)
df2['n.sh'] = df2['sh'].map(dict_2)

#Creates factor plot
sns.factorplot(x= 'n.esp', y = 'n.sh', data=df2)
plt.xlabel("End Semester Percentage")
plt.ylabel("Study Hours")
plt.yticks(ticks = [1,2,3], labels = ["(< 2 Hours)","(>= 4 Hours)", "(>= 6 Hours)"])
plt.show()

#4 Get the correlation from the data, then plug it to heatmap function and show it as following:
print("\n4. Correlation Heatmap")
#Get the Correction - Used Pearson method due to the large amount of categorical varaibles
corr = df1.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)

#Display of the numeric values of the correlation
print("Correlation Values")
print(corr)

#Plug correlation values into a heatmap
print("Correlation Heatmap")
sns.heatmap(corr)
plt.show()
```
